```python
# library imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# set matplotlib & seaborn theme
sns.set_theme(style="white")
```

```python
shelter = pd.read_csv(
    "https://raw.githubusercontent.com/giovani-gutierrez/animal_shelter_outcomes/refs/heads/main/data/raw.csv"
)
```

```python
# function that converts string from pascal to snake case
def pascal_to_snake(name):
    # insert underscore before each uppercase letter that follows a lowercase letter
    snake = re.sub(r"(?<=[a-z])(?=[A-Z])", "_", name)
    return snake.lower()


# convert column names from pascal to snake case
df = shelter.copy()
df.columns = [pascal_to_snake(col) for col in df.columns]

# display converted column names
df.columns
```

```python
# drop column with zero non-null values
df = df.drop(columns=["outcome_subtype"])
```

```python
date_cols = ["dob", "intake_date", "outcome_date", "last_update"]

for col in date_cols:
    df[col] = pd.to_datetime(df[col])
```

```python
# list of categorical columns
cat_cols = df.select_dtypes(include=["object"]).columns

# lowercase all values of each categorical column
df[cat_cols] = df[cat_cols].apply(lambda x: x.str.lower().str.strip())
```

```python
# drop any duplicate rows
df = df.drop_duplicates()
```

```python
# verify dtypes
df.dtypes
```

```python
# save cleaned data to parquet file
df.to_parquet("cleaned_data.parquet")
```
